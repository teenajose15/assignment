```python
# to remove outliers using IQR
def remove_outliers_iqr(df):
    Q1 = df.quantile(0.25)  
    Q3 = df.quantile(0.75)  
    IQR = Q3 - Q1           

    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the dataset to keep only non-outliers
    df_clean = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]
    
    return df_clean


data = remove_outliers_iqr(df)

data.shape
```

```python
# Number of features
num_features = len(data.columns)
num_cols = 3  # Adjust number of columns per row
num_rows = (num_features // num_cols) + (num_features % num_cols > 0)

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten for easy iteration

# Plot each feature separately
for i, col in enumerate(data.columns):
    axes[i].boxplot(data[col], vert=True)
    axes[i].set_title(col)
    axes[i].grid(True)

# Hide unused subplots
for i in range(num_features, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_ds = scaler.fit_transform(data)
     
# Creating a DataFrame with the scaled values
scaled_df = pd.DataFrame(scaled_ds, columns=data.columns)
scaled_df
```

# 1.1 kmeans Model

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
 kmeans=KMeans(n_clusters=i,init="k-means++",random_state=42)
 kmeans.fit(scaled_df)
 wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("No of Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++",random_state=42)
kmeans.fit(scaled_df)
y_kmeans=kmeans.fit_predict(scaled_df)
y_kmeans
```

```python
from sklearn.metrics import silhouette_score
sil_kmeans = silhouette_score(scaled_df, y_kmeans)
sil_kmeans
     
```

```python
# Scatter plot of clusters
plt.scatter(scaled_df['Alcohol'], scaled_df['Flavanoids'], s=30, c=y_kmeans, cmap='viridis')
plt.show()
```

# 1.2 Kmean using PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_df=pca.fit_transform(scaled_df)
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++",random_state=42)
kmeans.fit(pca_df)
y_kmeans_pca=kmeans.fit_predict(pca_df)
y_kmeans_pca
     
```

```python
sil_kmeans_pca = silhouette_score(pca_df, y_kmeans_pca)
sil_kmeans_pca
```

# 2.1 Agglomerative

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(scaled_df, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ahc=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
y_ahc=ahc.fit_predict(scaled_df)
y_ahc
```

```python
from sklearn.metrics import silhouette_score
sil_ahc=silhouette_score(scaled_df,y_ahc)
sil_ahc
```

```python
# Scatter plot of clusters
plt.scatter(scaled_df['Alcohol'], scaled_df['Flavanoids'], s=30, c=y_ahc, cmap='viridis')
plt.show()
```

# 2.2 Agglomerative with pca
